```python
#Lagrange Interpolation
import numpy as np
n=int(input("Enter number of data points:"))
x=np.zeros((n))
y=np.zeros((n))
print("Enter data for x and y:")
for i in range(n):
  x[i]=float(input('x['+str(i)+']='))
  y[i]=float(input('y['+str(i)+']='))
xp=float(input("Enter interpolation point:"))
yp=0
for i in range(n):
  p=1
  for j in range(n):
    if i!=j:
      p=p*(xp-x[j])/(x[i]-x[j])
  yp=yp+p*y[i]
print("Interpolated value at %.3f is %.3f."%(xp,yp))
```

```python
#Plotting single point on a plane
import matplotlib.pyplot as plt
x=2
y=4
plt.plot(x,y,'bo')
plt.grid()
plt.axis([-8,8,-8,8])
plt.title('Single Point Plot')
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
time=np.arange(-3*np.pi,3*np.pi,0.01)
amplitude=np.sin(time)
plt.plot(time,amplitude)
plt.title("sine wave",color='b')
plt.xlabel('Time'+r'$\rightarrow$')
plt.ylabel('Sin(time)'+r'$\rightarrow$')
plt.grid()
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
time=np.arange(-3*np.pi,3*np.pi,0.01)
amplitude=np.cos(time)
plt.plot(time,amplitude)
plt.title("Cos wave",color='b')
plt.xlabel('Time'+r'$\rightarrow$')
plt.ylabel('Cos(time)'+r'$\rightarrow$')
plt.grid()
plt.axhline(y=0,color='k')
plt.axvline(x=0,color='k')
plt.show()
```
